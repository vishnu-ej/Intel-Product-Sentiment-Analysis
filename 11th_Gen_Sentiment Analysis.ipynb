```python
import glob

# list all csv files only
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
import pandas as pd

df_csv_append = pd.DataFrame()

# append the CSV files
for file in csv_files:
    df = pd.read_csv(file)
    df_csv_append = pd.concat([df_csv_append, df], ignore_index=True)

df_csv_append
```

```python
import pandas as pd

# Load the CSV file
df = df_csv_append

df['text'] = df['text'].astype(str)

# Assuming df_csv_append is your DataFrame with the 'text' column converted to string
df_csv_append.to_csv('df_csv_append.csv', index=False)
```

```python
from transformers import DistilBertTokenizer

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

def preprocess_text(text):
    return tokenizer(text, padding=True, truncation=True, return_tensors='pt')

# Tokenize the text
tokens = [preprocess_text(text) for text in df['text']]
```

```python
from transformers import DistilBertForSequenceClassification

model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased-finetuned-sst-2-english')
```

```python
import torch

def predict_sentiment(tokenized_input):
    with torch.no_grad():
        outputs = model(**tokenized_input)
        logits = outputs.logits
        predicted_class = torch.argmax(logits, dim=1).item()
        return predicted_class

# Predict sentiment for all reviews
df['sentiment'] = [predict_sentiment(tokens[i]) for i in range(len(tokens))]
```

```python
df = df.drop(df.columns[2], axis=1)
df.to_csv('11thgenreviews_with_sentiment.csv', index=False)
```

```python
import pandas as pd

# Load the DataFrame from the CSV file
df_new = pd.read_csv('11thgenreviews_with_sentiment.csv')

# Extract the numerical part of the rating
df_new['rating'] = df_new['rating'].str.extract(r'(\d\.\d)').astype(float)

# Count the number of each rating
rating_counts = df_new['rating'].value_counts().sort_index()

# Print the counts of each rating
print("Ratings Count:")
for rating in [5.0, 4.0, 3.0, 2.0, 1.0]:
    print(f"{rating}: {rating_counts.get(rating, 0)}")

# Count the number of each sentiment
sentiment_counts = df_new['sentiment'].value_counts()

# Print the counts of each sentiment (0 and 1)
print("\nSentiment Count:")
print(f"0: {sentiment_counts.get(0, 0)}")
print(f"1: {sentiment_counts.get(1, 0)}")
```
